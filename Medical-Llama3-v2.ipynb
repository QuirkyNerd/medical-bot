```python
import requests

GROQ_API_KEY = ""
GROQ_API_KEY = "https://api.groq.com/openai/v1/chat/completions"
```

```python
def askme(question):
    system_message = """
You are an advanced Medical AI Assistant.

Provide responses in the following structured format:
1. Possible Causes
2. Symptoms Explanation
3. Recommended Actions
4. When to Consult a Doctor

Ensure accuracy, clarity, and safety. Avoid speculation.
"""

    headers = {
        "Authorization": f"Bearer {GROQ_API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.3-70b-versatile",
        "messages": [
            {"role": "system", "content": system_message},
            {"role": "user", "content": question}
        ],
        "temperature": 0.7,
        "max_tokens": 500,
        "top_p": 0.9
    }

    response = requests.post(GROQ_API_KEY, headers=headers, json=data)

    if response.status_code != 200:
        return f"Error {response.status_code}: {response.text}"

    reply = response.json()['choices'][0]['message']['content']

    disclaimer = "\n\nDisclaimer: This information is for educational purposes only and does not substitute professional medical advice. Please consult a qualified healthcare provider."

    return reply.strip() + disclaimer
```

```python
question = """I'm a 35-year-old male and for the past few months, I've been experiencing fatigue,
increased sensitivity to cold, and dry, itchy skin.
Could these symptoms be related to hypothyroidism?
If so, what steps should I take to get a proper diagnosis and discuss treatment options?"""

print(askme(question))
```
